```python
import numpy as np
import scipy as sci
import os
import pandas as pd
import shutil
import copy
```

```python
df_init = pd.read_csv('species_benzene_benzyl_alcohol.csv')
df = df_init
df
```

```python
paths_g = df['path_l_geom'].to_list()
paths_f = df['path_l_freq'].to_list()
calcs = df['Calc'].to_list()

for i in range(len(df)):
    directory_path = calcs[i]
    if not os.path.exists(directory_path):
        os.makedirs(directory_path, exist_ok=True)
    
    new_path = directory_path
    #geoms
    file = '/OUT.out'
    shutil.copy2(paths_g[i]+file,new_path+'/geom_opt.out')
    
    #file = '/volume.txt'
    #shutil.copy2(paths_g[i]+file,new_path+file)
    
    file = '/OUT.out'
    shutil.copy2(paths_f[i]+file,new_path+'/freq.out')
    
    #freqs
    
    file = '/Energy.txt'
    shutil.copy2(paths_g[i]+file,new_path+file)
    
    file = '/G_cds.txt'
    shutil.copy2(paths_g[i]+file,new_path+file)
    
    file = '/Frequencies.txt'
    shutil.copy2(paths_f[i]+file,new_path+file)
    
    file = '/inertia.txt'
    shutil.copy2(paths_f[i]+file,new_path+file)
    
    file = '/H_out.txt'
    shutil.copy2(paths_f[i]+file,new_path+file)
    
    file = '/S_out.txt'
    shutil.copy2(paths_f[i]+file,new_path+file)
    
    file = '/out_head.txt'
    shutil.copy2(paths_g[i]+file,new_path+file)
    #geom files:
    # image1.png
    # image2.png
    # OUT.out
    # volume.txt
    
    #freq files
    # OUT.out
```
